```python
import numpy as np
import matplotlib.pyplot as plt
import henon_map as hm
```

```python
n_samples = 1000
x = np.linspace(0, 1, n_samples)
xx, yy = np.meshgrid(x,x)
x_f = xx.flatten()
y_f = yy.flatten()
```

```python
engine = hm.partial_track.generate_instance(x_f, np.zeros_like(x_f), y_f, np.zeros_like(x_f))
```

```python
engine
```

```python
data = engine.compute(10000, 0.0)
```

```python
plt.figure()

plt.imshow(data[4].reshape(n_samples, n_samples))
plt.colorbar()
```

```python
engine.reset()
```

```python
data = engine.compute(10000, 64.0)
```

```python
plt.figure()

plt.imshow(data[4].reshape(n_samples, n_samples))
plt.colorbar()
```

```python
engine.reset()
```

```python
data = engine.compute(10000, 16.0, mu=0.2)
```

```python
plt.figure()

plt.imshow(data[4].reshape(n_samples, n_samples))
plt.colorbar()
```
